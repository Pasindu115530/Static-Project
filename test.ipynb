```python
from scipy.stats import ttest_1samp
import numpy as np

# Make sure 'calculated_total' exists
if "calculated_total" not in df.columns:
    expense_cols_existing = [col for col in expense_cols if col in df.columns]
    df["calculated_total"] = df[expense_cols_existing].sum(axis=1)

# Data
data = df["calculated_total"].dropna()

# Hypothetical mean
hypothetical_mean = 15000  # example value in LKR

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, hypothetical_mean)

print(f"One-sample t-test against mean = {hypothetical_mean} LKR")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:}")

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis → mean is significantly different from 15,000 LKR")
    print("Interpretation: The average total monthly expenditure of students is statistically different from 15,000 LKR.")
else:
    print("Fail to reject null hypothesis → no significant difference from 15,000 LKR")
    print("Interpretation: The average total monthly expenditure of students is not statistically different from 15,000 LKR.")
```

```python
# Select the two groups by gender
group_male = df[df["gender"] == "Male Student"]["calculated_total"].dropna()
group_female = df[df["gender"] == "Female Student"]["calculated_total"].dropna()

# Perform t-test (Welch’s t-test)
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(group_male, group_female, equal_var=False)

print("T-test: Male vs Female Total Expenditure")
print(f"T-statistic: {t_stat:}")
print(f"P-value: {p_value:}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis → mean expenditures are significantly different")
    print("Interpretation: There is a significant difference in average total monthly expenditure between male and female students.")
else:
    print("Fail to reject null hypothesis → no significant difference")
    print("Interpretation: There is no statistically significant difference in average total monthly expenditure between male and female students.")
```

```python
# Define the boarding/hostel categories
boarding_categories = ["Boarding place", "Rented apartment/room", "On-campus hostel"]

# Select the two groups
group_boarding = df[df["living_arrangement"].isin(boarding_categories)]["calculated_total"].dropna()
group_home = df[df["living_arrangement"] == "Living with family"]["calculated_total"].dropna()

# Perform t-test (Welch’s t-test)
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(group_boarding, group_home, equal_var=False)

print("T-test: Boarding/Hostel vs From Home Total Expenditure")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis → mean expenditures are significantly different")
    print("Interpretation: Students living in boarding/hostel arrangements spend significantly more or less on average than students living from home.")
else:
    print("Fail to reject null hypothesis → no significant difference")
    print("Interpretation: There is no statistically significant difference in average total expenditures between students living in boarding/hostel arrangements and students living from home.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Select numeric expense columns
expense_cols_existing = [col for col in expense_cols if col in df.columns]

# Ensure 'calculated_total' exists
if "calculated_total" not in df.columns:
    df["calculated_total"] = df[expense_cols_existing].sum(axis=1)

# Features (X) = all individual expenses
X = df[expense_cols_existing]
y = df["calculated_total"]

# Drop rows with missing values
X = X.dropna()
y = y[X.index]

n_samples = len(y)
if n_samples < 2:
    print(f"Dataset too small to compute R² or cross-validation (only {n_samples} sample(s)).")
else:
    # Initialize model
    model = LinearRegression()
    
    # Fit on all data
    model.fit(X, y)

    # Predict total expenditure
    y_pred = model.predict(X)

    # Show coefficients
    coefficients = pd.DataFrame({
        "Expense": expense_cols_existing,
        "Coefficient": model.coef_
    })
    print("\nRegression Coefficients:")
    display(coefficients)

    # Plot predicted vs actual
    plt.figure(figsize=(8,6))
    plt.scatter(y, y_pred, alpha=0.7)
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2)
    plt.xlabel("Actual Total Expenditure")
    plt.ylabel("Predicted Total Expenditure")
    plt.title("Predicted vs Actual Total Expenditure")
    plt.show()
```

1. Total Monthly Expenditure

Calculated as the sum of all individual expenses (rent, utilities, groceries, canteen food, outside food, transport, stationery, printing, personal care, clothing, gym, medical costs).

Grand statistics:

Mean: Indicates the average monthly expenditure of students.

Median: Represents the typical student’s expenditure, less affected by extreme values.

Mode: Most frequently occurring expenditure range.

Standard Deviation: Shows the variability among students’ spending.

2. Expense Distribution

Food Expenses (canteen vs outside): Overlapping histograms show that most students spend more on outside food compared to canteen food.

Other Expenses (utilities, stationery, printing, personal care, clothing): Histograms reveal spread and skewness.

Positive skew indicates a few students spend much more than average.

3. Skewness and Variability

Skewness indicates whether the distribution is symmetric or skewed.

Standard deviation identifies which expense categories vary the most.

Example: Outside food or rent may show higher variation compared to stationery.

4. Group Comparisons

Gender: T-test indicates whether male and female students have statistically different total expenditures.

Living Arrangement: T-test compares students living at home vs hostel/boarding.

Significant p-values indicate different spending patterns between groups.

5. Correlation Analysis

Positive correlations between related categories (e.g., groceries and outside food, utilities and rent).

Strong correlation with total expenditure identifies major contributing expenses.

Heatmap visualizations help spot patterns quickly.

6. Predictive Modeling

Linear regression shows how each expense category contributes to total monthly expenditure.

Coefficients indicate the influence of each category:

Higher coefficients → larger impact on total expenditure.

Predicted vs actual plots indicate the model’s ability to estimate total expenditure.

7. Key Insights

Rent, outside food, and utilities are the largest contributors to total expenditure.

Some categories (stationery, printing) have low impact individually but still contribute to overall variability.

Students living away from home or in hostels tend to have higher expenses.

Gender differences may exist in certain categories but require statistical confirmation

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, mode

# Safe mode function
def safe_mode(series):
    if len(series) == 0:
        return None
    m = mode(series, keepdims=True)
    if hasattr(m, "mode") and len(m.mode) > 0:
        return m.mode[0]
    else:
        return m[0] if isinstance(m, (list, np.ndarray)) else m

# Select numeric expense columns
expense_cols_existing = [col for col in expense_cols if col in df.columns]

# Ensure 'calculated_total' exists
if "calculated_total" not in df.columns:
    df["calculated_total"] = df[expense_cols_existing].sum(axis=1)

# Include calculated_total in analysis
all_cols = expense_cols_existing + ["calculated_total"]

# Initialize summary dictionary
summary_dict = {
    "Mean": [],
    "Median": [],
    "Mode": [],
    "Std Dev": [],
    "Skewness": [],
    "Correlation with Total": []
}

for col in all_cols:
    data = df[col].dropna()
    summary_dict["Mean"].append(data.mean())
    summary_dict["Median"].append(data.median())
    summary_dict["Mode"].append(safe_mode(data))
    summary_dict["Std Dev"].append(data.std())
    summary_dict["Skewness"].append(skew(data))
    summary_dict["Correlation with Total"].append(data.corr(df["calculated_total"]))

# Create summary DataFrame
summary_df = pd.DataFrame(summary_dict, index=all_cols)

# Display the table
display(summary_df)
```
